```python
nodes_dframe=pd.read_csv(nodes_file)
edges_dframe=pd.read_csv(edges_file)
hierarch_dframe=pd.read_csv(hierarch_file, index_col=1)
```

```python
frontal_nodes=pd.read_csv(all_frontal_nodes_file)
flat_frontal_nodes_wM1=pd.read_csv(flat_frontal_nodes_withM1_file)
```

```python
nodelist=nodes_dframe.id.tolist()
```

##Build whole brain graph

```python
edgelist=[]
[edgelist.append(x) for x in edges_dframe.values]
num_ms_digraph=nx.DiGraph(edgelist)
```

add in hierarchy as node attribute

```python
for x in num_ms_digraph.nodes_iter():
    num_ms_digraph.node[x]['level']=hierarch_dframe.ix[x][0]
```

make undirected

```python
num_ms_graph=num_ms_digraph.to_undirected()
```

##Relabel nodes with names of regions rather than numbers

```python
zipped_nodes=zip(nodes_dframe.id.values, nodes_dframe.name.values)
nodes_dict=dict(zipped_nodes)
```

```python
ms_graph=nx.relabel_nodes(num_ms_graph,nodes_dict)
```

##Build frontal subgraph

```python
frontal_nodelist=list(frontal_nodes.name)
frontal_graph=ms_graph.subgraph(frontal_nodelist)
```

```python
remove_list=[]
[remove_list.append(x[0]) for x in frontal_graph.degree_iter() if x[1]==0]
```
